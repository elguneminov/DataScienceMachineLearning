# Seaborn

```python
import seaborn as sns
%matplotlib inline
```

```python
data = sns.load_dataset('tips')
```

```python
sns.distplot(data['total_bill'], kde = False, bins = 30)
```

```python
# { "scatter" | "reg" | "resid" | "kde" | "hex" }
sns.jointplot(x = 'total_bill', y = 'tip', data = data, kind = 'hex')
```

```python
sns.pairplot(data, hue = 'sex', palette = 'coolwarm' )
```

```python
data.head()
```

```python
sns.rugplot(data['total_bill'])
sns.kdeplot(data['total_bill'])
```

```python
sns.rugplot(data['tip'])
sns.kdeplot(data['tip'])
```
