```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

1. factorplot
2. boxplot
3. violinplot
4. stripplot
5. swarmplot
6. barplot
7. countplot

```python
sns.barplot(x = 'time', y = 'total_bill', data = tips)
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips)
```

```python
import numpy as np
```

```python
sns.barplot(x = 'smoker', y = 'total_bill', data = tips, estimator=np.median)
```

```python
sns.countplot(y = 'sex', data = tips, hue = 'time')
```

```python
sns.countplot(x = 'sex', data = tips, hue = 'day')
```

```python
tips['sex'].value_counts()
```

```python
sns.boxplot(x = 'sex', y = 'total_bill', data= tips, palette='rainbow', hue = 'smoker')
```

```python
tips.head()
```

```python
sns.violinplot(x = 'smoker', y = 'total_bill', data=tips, hue = 'sex')
```

```python
sns.stripplot(x = 'sex', y = 'tip', data=tips, hue = 'smoker')
```

```python
sns.swarmplot(x = 'day', y = 'tip', data=tips, hue = 'smoker')
```

```python
sns.factorplot(x = 'day', y = 'tip', data=tips, kind = 'bar')
```
